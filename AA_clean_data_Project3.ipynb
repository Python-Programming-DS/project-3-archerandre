```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load in data
df = pd.read_csv('new_train_EGN5442.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(), df.nunique()
```

```python
#We need to remove the unnamed column (the first one), which is just an index
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
#there are a lot of values that haev strings involved with numbers, lets strip those and make them into just numbers. 
#This is x2, x29
#x2 has the format of #123 when it should just be 123
#x29 has the format (123&&) when it shoudl jsut be 123
df['x2'] = df['x2'].str.replace('#','').astype(float)
df['x29'] = df['x29'].str.extract(r'\((\d+)&&\)').astype(float)
df.head()
```

```python
#convert x3 to 0 and 1. If Female, then 0, if Male, then 1
df['x3'] = df['x3'].map({'Female':0, 'Male':1})
df.head()
```

```python
#Lets see how many unique values are in the x25 column
df['x25'].nunique(), df['x25'].unique()
#Set D_C to 0, and L_C to 1
df['x25'] = df['x25'].map({'D_C':0, 'L_C':1})
df.head()
```

```python
#Now do the same check for x26
df['x26'].nunique(), df['x26'].unique()
# Set Bt to 0, PT to 1
df['x26'] = df['x26'].map({'Bt':0, 'PT':1})
df.head()
```

```python
#Is the x30 column all empty?
print(df['x30'].isnull().sum(), len(df['x30']))
#Yes, so we can drop it
df.drop(columns=['x30'], inplace=True)
df.head()
```

```python
#x4 is all one number, so we can drop it
print(df['x4'].nunique(), df['x4'].unique())
df.drop(columns=['x4'], inplace=True)
df.head()
```

```python
# it is really really difficult to know the best way to handle the inf values in x31 wihtout any understanding of the dataset. SO lets convert to NaN and see how much data is missing
df['x31'] = df['x31'].replace([np.inf, -np.inf], np.nan)
print(df['x31'].isnull().sum()/len(df['x31']) * 100, "% of data is missing in x31")

#Since over half of the data is missing in x31, we will drop this column
df.drop(columns=['x31'], inplace=True)
df.head()
```

```python
#now lets see if there are any missing values in the dataset, there are only msising values in x24
print( df.isnull().sum() )
#drop rows with missing values
df.dropna(inplace=True)
print( df.isnull().sum() )
```

```python
df.info(), df.describe(), df.nunique()
```

```python
#x14 is all strings for some reason, we need to change the datatypes to numeric
#Lets see what kind of strings are in there
print(df['x14'].unique())
#they are all numbers or " ", so we can just replace the " " with NaN and convert to float
df['x14'] = df['x14'].replace(' ', np.nan).astype(float)
#Now drop any rows with NaN in x14
df.dropna(subset=['x14'], inplace=True)
df.info()
```

```python
#Now this data is cleaned! We can save it to a new csv
df.to_csv('cleaned_train_EGN5442.csv', index=False)
```
